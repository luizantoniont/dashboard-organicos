```python
import pandas as pd

URL_ROOT = r'https://www.gov.br/agricultura/pt-br/assuntos/sustentabilidade/organicos/CNPO_MAPA_03_07_2024_IMASCARA.xlsx'
data = pd.read_excel(URL_ROOT, header=2)
```

```python
import re

def formart_columns_name(columns_name):
    columns_name = re.sub(' ', '_', columns_name)
    return columns_name.lower()

data.columns = [formart_columns_name(column) for column in data.columns]

data.columns
```

```python
import sqlite3

banco_dados = 'cadastro_produtores_organicos.db'
tabela_nome = 'dados_excel'

def salvar_sqlite(data, banco_dados, tabela_nome):
    connection = sqlite3.connect(banco_dados)
    cursor = connection.cursor

    data.to_sql(tabela_nome, connection, if_exists='replace', index='false')

    
    connection.close()

salvar_sqlite(data, banco_dados, tabela_nome)

```
